```python
<h3>Use a classification binary cutoff based on Average Employed Rate</h3>
```

```python
#https://www.geeksforgeeks.org/ways-to-apply-an-if-condition-in-pandas-dataframe/

df2.loc[df2["Avg Employed Rate"] >= 62, 'HighEmp'] = 'True'
df2.loc[df2["Avg Employed Rate"] < 62, 'HighEmp'] = 'False'
df2
```

```python
df2["HighEmp"].describe()
```

```python
from pycaret.classification import *
```

```python
#Best to use datetime.strptime here
train=df2[1:150]
test=df2[151:]
train
```

```python
s = setup(data = train, test_data = test, ignore_features=['Avg Employed Rate'],
          target = 'HighEmp', fold_strategy = 'timeseries', fold = 3, session_id = 124)
```

```python
#https://towardsdatascience.com/time-series-forecasting-with-pycaret-regression-module-237b703a0c63
best = compare_models()
```

```python
lr=create_model(best)
prediction_holdout = predict_model(lr)
```

<h3>About 85% accuracy on training data, 75% accuracy on test</h3>

```python
%matplotlib inline
evaluate_model(lr)
```
