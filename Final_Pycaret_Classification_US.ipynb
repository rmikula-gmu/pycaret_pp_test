<h3>Use a classification binary cutoff based on Average Employed Rate</h3>

```python
#https://www.geeksforgeeks.org/ways-to-apply-an-if-condition-in-pandas-dataframe/

df2.loc[df2["Avg Percent of Population"] >= 62, 'HighEmp'] = 'True'
df2.loc[df2["Avg Percent of Population"] < 62, 'HighEmp'] = 'False'
df2
```

```python
df2["HighEmp"].describe()
```

```python
from pycaret.classification import *
```

```python
#Best to use datetime.strptime here
train=df2[1:714]
test=df2[715:]
train
```

```python
s = setup(data = train, test_data = test, ignore_features=['Avg Percent of Population'],
          target = 'HighEmp', fold_strategy = 'timeseries', fold = 3, session_id = 124)
```

```python
#https://towardsdatascience.com/time-series-forecasting-with-pycaret-regression-module-237b703a0c63
best = compare_models()
```

```python
ada=create_model("ada")
prediction_holdout = predict_model(ada)
```

<h3>About 91% accuracy on training data, 96% accuracy on test fior ada boost</h3>

```python
%matplotlib inline
evaluate_model(ada)
```
